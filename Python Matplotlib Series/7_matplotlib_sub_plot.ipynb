```python
# display the multiple plot - with subplot() you can draw multiple plots in one fighure
import matplotlib.pyplot as plt
import numpy as np

# plot1
x = np.array([0,1,2,3])
y = np.array([3,8,1,10])
plt.subplot(1,2,1) # the figure has 1 row, 2 columns and this plot is the 1 plot.
plt.plot(x,y)

# plot2
x = np.array([0,1,2,3])
y = np.array([10,20,30,40])
plt.subplot(1,2,2)

plt.show()
```
