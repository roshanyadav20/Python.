```python
# Histogram- A histogram is a graph showing the frequency distribution
# ex- say you ask for the height of 250 people then how we will make a histogram.
# hist() function

import numpy as np
x = np.random.normal(170, 10, 250)
print(x)
```

```python
# the hist() function will read the array and produce the histogram.
import matplotlib.pyplot as plt
import numpy as np

x = np.random.normal(170, 10, 250)
plt.hist(x)
plt.show()
```
