Arithmetic Operation

```python
import pandas as pd 

var = pd.DataFrame({"A":[1,2,3,4],"B":[5,6,7,8]})

var
```

```python
# add
var["C"] = var["A"]+var["B"]
var
```

```python
# subtract
var["C"] = var["A"]-var["B"]
var
```

```python
# multiply
var["C"] = var["A"]*var["B"]
var
```

```python
# devide
var["C"] = var["A"]/var["B"]
var
```
