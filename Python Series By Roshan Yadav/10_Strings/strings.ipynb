Strings in Python

- A string is a sequence of characters.
- strings-> single, double and triple quotes

```python
name = "Roshan"  # create a string
print(name)
print(type(name))  #checking data type
```

```python
print("Hello World")
print("It's easy")
print("\"Hello World\"")
```

Type 1: Formatted Strings

```python
# 1. Old styple formatting - % operator
# %s, %d are placeholders for the string and integer.

name = "Madhav"
age = 16
print("My name is %s and I'm %d" % (name, age))
```

Type 2: Formatted String - str.format()

```python
# str.format() method
name = "Roshan"
age = 16
print("My name is {} and I'm {}".format(name, age))
```

```python
name = "Roshan"
age = 16
print("My name is {0} and I'm {1}".format(name, age))

# you can reference variables by index or keyword
print("My name is {0} and I'm {1}".format(name, age))
print("My name is {1} and I'm {0}".format(name, age))

print("My name is {name} and I'm {age}".format(name="Roshan", age = "22"))

```

Type 3: Formatted String - F-strings

```python
# f-strings
# Mera favourite hai.

name = "Roshan"
age = 24
print(f"My name is {name} and I'm {age}")

print(f"My age after 5 years will be {age + 5}")
```

String Operators

```python
# String Operators in Python-
a = "Hello"
b = "Python"

print(a+b) # concatenate
print(a*2) # multiple copies
```

```python
# in operator

a = "Hello"
if "H" in a:
    print("Yess")
else:
    print("no")
        
```

```python
# not in operator

if "h" not in a:
    print("Yes")
else:
    print("no")    
```
