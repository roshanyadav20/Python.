Arguments in Functions

1. Required Arguments (single/multiple arguments)

```python
def greetings(name):  # name is a parameter
    print("Hello ", name, "!")

greetings("Roshan") # Roshan is argument
# greetings()  # required an argument to run code. nhi to error ayega.  
```

```python
def intro(course_name, instructor_name):
    print("Welcome to ", course_name, "course by", instructor_name)
    
intro("Python", "Roshan")    
```

2. Default Arguments

```python
def greetings(name = "World"):
    print("Hello", name, "!")
    
greetings() # runs without error using default value
greetings("Roshan")    
```

3. Keyword Arguments (named argument)

```python
def divide(a,b):
     return a/b

result1 = divide(100,20) #positional argument
print(result1)

result2 = divide(100,20) # positional argument
print(result2)

result3 = divide(b=100, a=20) #keyword argument
print(result3)
```

4. Arbitrary Positional Arguments(*args)

```python
# 1- jab apko pta hi na ho ki kitne no of arguments function me pass krne ho, then use.
# 2- jo variable ki value matlab arguments pass kreneg wo as a tuple store hoga.

# fix argument pass
def add2number(a,b):
    return a+b

result = add2number(10,11)
print(result)
```

```python
# fix argument pass
def add2number(a,b,c):
    return a+b+c

result = add2number(10,11,12)
print(result)
```

```python
# Arbitary use
# Stores argument as a tuple
def add_numbers(*args):
    print(type(args))
    return sum(args)

output1 = add_numbers(1,2)
output2 = add_numbers(1,2,3,4) # variable no ofarguments
print(output1)
print(output2)
```

```python
def greetings2(*names):
    for name in names:
       print(f"Hello, {name}!")
       
greetings2("Madhav", 'Roshan', 'Rohan')        
    
```

5. Arbitrary Keyword Arguments (**kwargs)

```python
# multiple keyword arguments pass kr sakte hai.
# stores arguments as dictionary

def print_details(**kwargs):
    for key, value in kwargs.items():
        print(f"{key}: {value}")
        
print_details(name = "Roshan", age = 26, city ="Sultanpur")       
    
```

```python
def print_details(**kwargs):
    print(type(kwargs))
    for key, value in kwargs.items():
        print(f"{key}: {value}")
        
print_details(name = "Roshan", age = 26, city ="Sultanpur")  
```
