- Camel Case

- snake case

Rules to declare variables:-

```python
# 1. Variable name must start with: letter or _

_my_name = "Roshan"  
print(_my_name)

myname = "Roshan"
print(myname)
```

```python
# 2. Variable name can contain: letter, numbers, underscore (_)

my_name1 = "Roshan"
r1 = 26
_1 = 1
```
